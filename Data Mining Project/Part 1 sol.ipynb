```python
#processing input1 , input2 and writting in input3 for further processing. 
data = data2 = "" 
  
# Reading data from file1 
with open('input1.txt') as fp: 
    data = fp.read() 
  

with open('input2.txt') as fp: 
    data2 = fp.read() 
  
 
data += "\n"
data += data2 
  
with open ('input3.txt', 'w') as fp: 
    fp.write(data) 
```

```python
from collections import Counter
def word_count(fname):
        with open(fname) as f:
                return Counter(f.read().split())

print("The number of words with a specific amount of letters in both input file :",word_count("input3.txt"))
```

```python
import re
import string
frequency = {}
document_text = open('input3.txt', 'r')
text_string = document_text.read().lower()
match_pattern = re.findall(r'\b[a-z]{3,15}\b', text_string)
 
for word in match_pattern:
    count = frequency.get(word,0)
    frequency[word] = count + 1
     
frequency_list = frequency.keys()
 
for words in frequency_list:
    print (words, frequency[words])
```
